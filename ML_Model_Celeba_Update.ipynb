```python
# Import Required Library For ML Prediction

import numpy as np
import pandas as pd
```

```python
# Import Total Dataset Of Celeba For ML Prediction

dataset = pd.read_csv('celeba_total_dataset_renew.csv')
```

```python
# Visualisizing Dataset For ML Prediction (Head)

dataset.head(5)
```

```python
# Visualisizing Dataset For ML Prediction (Tail)

dataset.tail(5)
```

```python
# set Dependent And Independent Value Of Dataset In Two Var Named :- x , y

x = dataset.iloc[0:2000, 1:55]
y = dataset.iloc[0:2000, 55:56]
```

```python
# Visualizing Imported Dataset Of x

x
```

```python
# Visualizing Imported Dataset Of y

y
```

```python
# Wit The Help Of Sk-learn Lab Convert Strings Into Number Formate

#from sklearn.preprocessing import LabelEncoder
#label_encoder_y = LabelEncoder()
#y = label_encoder_y.fit_transform(y)
```

```python
# Visualizing Imported Dataset Of y After Encoding String To Number Formate

y
```

```python
# With The HElp of Sk-learn Lab Train & Test The Model 


from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x ,y , test_size = 0.29)
```

```python
# Applied Best preference ML Model (Linear Regression) A/q To Accuracy


from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(x_train,y_train)
```

```python
# Predict The Model After Applied ML Model

y_pred = lin_model.predict(x_test)

y_pred
```

```python
# Checking Accuracy In DataFrame Formate

Formate = pd.DataFrame(np.c_[(x_test) , (y_test) , (y_pred)])
Formate
```

## Model Accuracy

```python
# Check The Accurate Accuracy Of Built ML Model Using R2_score Method

from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

## Save The Model At Highest Accuracy 

```python
import joblib 
```

```python
joblib.dump(lin_model ,"Celeba_Update.joblib")
```
